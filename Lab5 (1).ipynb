```python
from qiskit import QuantumCircuit
```

```python
qc1 = QuantumCircuit(2)
qc1.cx(0, 1)
qc1.cx(1, 0)
print(qc1)
```

```python
qc2 = QuantumCircuit(3)
qc2.ccx(0, 1, 2)
qc2.cx(2, 1)
qc2.ccx(0, 1, 2)
print(qc2)
```

```python
qc3 = QuantumCircuit(2)
qc3.cx(0, 1)
qc3.cz(0, 1)
print(qc3)
```

```python
qc4 = QuantumCircuit(3)
qc4.ccx(0, 1, 2)
print(qc4)
```

```python
qc5 = QuantumCircuit(2)
qc5.swap(0, 1)
print(qc5)
```

```python
qc6 = QuantumCircuit(3)
qc6.cswap(0, 1, 2)
print(qc6)
```
