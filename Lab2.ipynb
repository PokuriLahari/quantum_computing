```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# Create 1 qubit and 1 classical bit
qc = QuantumCircuit(1, 1)

# Apply Hadamard gate (creates superposition)
qc.h(0)

# Measure qubit
qc.measure(0, 0)

# Run on simulator
simulator = AerSimulator()
job = simulator.run(qc, shots=1000)
result = job.result()

# Display qubit values
print("Qubit Measurement Output:")
print(result.get_counts())
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# Create quantum circuit: 1 qubit, 1 classical bit
qc = QuantumCircuit(1, 1)

# Apply Hadamard gate
qc.h(0)

# Measure qubit
qc.measure(0, 0)

# ----- DISPLAY CIRCUIT DIAGRAM -----
print("Quantum Circuit Diagram:")
print(qc.draw())   # TEXT diagram (NO ERROR)

# ----- RUN AND DISPLAY QUBIT VALUES -----
simulator = AerSimulator()
job = simulator.run(qc, shots=1000)
result = job.result()

print("\nQubit Measurement Values:")
print(result.get_counts())
```
