```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1, 1)

qc.h(0)
qc.x(0)
qc.y(0)
qc.z(0)

qc.measure(0, 0)

print("Single Qubit Circuit:")
print(qc.draw())
```

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2, 2)

qc.h(0)
qc.x(0)
qc.z(1)
qc.cx(0, 1)

qc.measure([0,1], [0,1])

print("Two Qubit Circuit:")
print(qc.draw())
```

```python
from qiskit import QuantumCircuit
import math

qc = QuantumCircuit(1, 1)

qc.h(0)
qc.rx(math.pi/2, 0)   # 90 degree correction
qc.rx(math.pi/4, 0)   # 45 degree correction

qc.measure(0, 0)

print("Single Qubit Error Correction Circuit:")
print(qc.draw())
```

```python
from qiskit import QuantumCircuit
import math

qc = QuantumCircuit(1, 1)

qc.x(0)                 # 180°
qc.rx(math.pi/2, 0)     # 90°
qc.rx(math.pi/4, 0)     # 45°

qc.measure(0, 0)

print("Single Qubit Pauli Angle Circuit:")
print(qc.draw())
```
